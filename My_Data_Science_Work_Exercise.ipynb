```python


print("========Simple Calculater==========")
print("====Operations +,-,*,/,**,//,% ====")
print("===================================")

# Step 1: Get inputs

num1 = float(input("Enter first number: "))
operater = input("Enter the operater (+,-,*,/,**,//,%):" )
num2 = float(input("Enter second number: "))

# Step 2: Perform calculation
if operater == "+":
  result = num1 + num2
elif operater == "-":
  result = num1 - num2
elif operater == "**":
  result = num1 ** num2
elif operater == "//":
  if num2 == 0:
    print("Error: Division by zero")
  else:
    result = num1 // num2
elif operater == "*":
  result = num1 * num2
elif operater == "/":
  if num2 == 0:
    print("Error: Division by zero")
  else:
    result = num1 / num2
elif operater == "%":
  result = num1 * num2

else:
    result = "Invalid operator!"

# Step 3: Show result
print(f"\nHistory of the last calculation: {num1} {operater} {num2} = {result}")
print("===================================")
```

```python
import random

def generate_secret(min_value, max_value):
  return random.randint(min_value, max_value)

def check_guess(secret, guess):
  """Check if guess is correct, too high, or too low"""
  if guess == secret:
      return "correct"
  elif guess < secret:
      return "too_low"
  else:
      return "too_high"


# main
def main():
  print("=" * 35)
  print("    NUMBER GUESSING GAME ")
  print("=" * 35)

  # condition

  min_value = 1
  max_value = 100
  max_attempts = 7

  # secret number
  secret = generate_secret(min_value, max_value)
  attempts = 0
  game_won = False

  print(f"I'm thinking of a number between {min_value} and {max_value}")
  print(f"You have {max_attempts} attempts. Good luck!\n")
  while attempts < max_attempts:
    attempts += 1
    remaining = max_attempts - attempts

# get player guess
    try:
      guess = int(input(f"Attempt {attempts}/{max_attempts} — Your guess: "))
    except ValueError:
      print("⚠️  Please enter a valid number!\n")
      attempts -= 1
      continue

# check the guess
    result = check_guess(secret, guess)

    if result == "correct":
      game_won = True
      break
    elif result == "too_low":
      print("⚠️  Too low!\n")
    else:
      print("⚠️  Too high!\n")


    if remaining > 0:
      print(f"You have {remaining} attempts remaining.\n")

# show result
  print ("=" * 35)

  if game_won:
    print("Congratulations! You guessed the secret number!")

    if attempts <= 3:
      print('Amazing! yours a genius')
    elif attempts <=5:
      print("Great job")
    else:
      print("you win")
  else:
    print("Game over! You ran out of attempts.")
    print(f"The number was {secret}")


  game_won == False
  if input("Do you want to play again? (yes/no): ").lower() == "yes":
    main()
  else:
    print("Thanks for playing! Goodbye!")
    game_won == True

main()
```
