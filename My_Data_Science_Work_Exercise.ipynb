```python


print("========Simple Calculater==========")
print("====Operations +,-,*,/,**,//,% ====")
print("===================================")

# Step 1: Get inputs

num1 = float(input("Enter first number: "))
operater = input("Enter the operater (+,-,*,/,**,//,%):" )
num2 = float(input("Enter second number: "))

# Step 2: Perform calculation
if operater == "+":
  result = num1 + num2
elif operater == "-":
  result = num1 - num2
elif operater == "**":
  result = num1 ** num2
elif operater == "//":
  if num2 == 0:
    print("Error: Division by zero")
  else:
    result = num1 // num2
elif operater == "*":
  result = num1 * num2
elif operater == "/":
  if num2 == 0:
    print("Error: Division by zero")
  else:
    result = num1 / num2
elif operater == "%":
  result = num1 * num2

else:
    result = "Invalid operator!"

# Step 3: Show result
print(f"\nHistory of the last calculation: {num1} {operater} {num2} = {result}")
print("===================================")
```

```python
import random

def generate_secret(min_value, max_value):
  return random.randint(min_value, max_value)

def check_guess(secret, guess):
  """Check if guess is correct, too high, or too low"""
  if guess == secret:
      return "correct"
  elif guess < secret:
      return "too_low"
  else:
      return "too_high"


# main
def main():
  print("=" * 35)
  print("    NUMBER GUESSING GAME ")
  print("=" * 35)

  # condition

  min_value = 1
  max_value = 100
  max_attempts = 7

  # secret number
  secret = generate_secret(min_value, max_value)
  attempts = 0
  game_won = False

  print(f"I'm thinking of a number between {min_value} and {max_value}")
  print(f"You have {max_attempts} attempts. Good luck!\n")
  while attempts < max_attempts:
    attempts += 1
    remaining = max_attempts - attempts

# get player guess
    try:
      guess = int(input(f"Attempt {attempts}/{max_attempts} — Your guess: "))
    except ValueError:
      print("⚠️  Please enter a valid number!\n")
      attempts -= 1
      continue

# check the guess
    result = check_guess(secret, guess)

    if result == "correct":
      game_won = True
      break
    elif result == "too_low":
      print("⚠️  Too low!\n")
    else:
      print("⚠️  Too high!\n")


    if remaining > 0:
      print(f"You have {remaining} attempts remaining.\n")

# show result
  print ("=" * 35)

  if game_won:
    print("Congratulations! You guessed the secret number!")

    if attempts <= 3:
      print('Amazing! yours a genius')
    elif attempts <=5:
      print("Great job")
    else:
      print("you win")
  else:
    print("Game over! You ran out of attempts.")
    print(f"The number was {secret}")


  game_won == False
  if input("Do you want to play again? (yes/no): ").lower() == "yes":
    main()
  else:
    print("Thanks for playing! Goodbye!")
    game_won == True

main()
```

```python
# =============================================
# PROJECT: Student Grade Tracker System
# =============================================

def calculate_average(scores):
    """Calculate average of a list of scores"""
    if len(scores) == 0:
        return 0
    return sum(scores) / len(scores)

def assign_grade(average):
    """Assign letter grade based on average"""
    if average >= 90:
        return "A+"
    elif average >= 80:
        return "A"
    elif average >= 70:
        return "B"
    elif average >= 60:
        return "C"
    else:
        return "F"

def create_student(name, scores):
    """Create a student dictionary"""
    avg   = calculate_average(scores)
    grade = assign_grade(avg)
    return {
        "name"    : name,
        "scores"  : scores,
        "average" : round(avg, 2),
        "grade"   : grade,
        "highest" : max(scores),
        "lowest"  : min(scores)
    }

def display_student(student):
    """Display one student's report"""
    print(f"\n  Student : {student['name']}")
    print(f"  Scores  : {student['scores']}")
    print(f"  Average : {student['average']}")
    print(f"  Grade   : {student['grade']}")
    print(f"  Highest : {student['highest']}")
    print(f"  Lowest  : {student['lowest']}")

def display_class_report(students):
    """Display full class analysis"""
    print("\n" + "=" * 45)
    print("          CLASS REPORT CARD ")
    print("=" * 45)

    for student in students:
        display_student(student)
        print("  " + "-" * 35)

    # Class statistics
    all_averages = [s["average"] for s in students]
    all_grades   = [s["grade"] for s in students]

    print("\n CLASS STATISTICS")
    print("-" * 35)
    print(f"  Total Students  : {len(students)}")
    print(f"  Class Average   : {calculate_average(all_averages):.2f}")
    print(f"  Highest Average : {max(all_averages)}")
    print(f"  Lowest Average  : {min(all_averages)}")

    # Grade distribution
    print("\n🏆 GRADE DISTRIBUTION")
    print("-" * 35)
    unique_grades = sorted(set(all_grades))
    for grade in unique_grades:
        count    = all_grades.count(grade)
        students_with_grade = [s["name"] for s in students
                               if s["grade"] == grade]
        print(f"  Grade {grade}: {count} student(s)"
              f" → {students_with_grade}")

    # Top performers
    top = [s for s in students if s["average"] >= 80]
    print(f"\n TOP PERFORMERS (avg ≥ 80): "
          f"{[s['name'] for s in top]}")

    # Students needing help
    struggling = [s for s in students if s["average"] < 60]
    if struggling:
        print(f"  NEEDS SUPPORT: "
              f"{[s['name'] for s in struggling]}")
    else:
        print(" No students failing!")


# ── Main Program ──────────────────────────────

# Student data (name, list of 5 subject scores)
data = [
    ("Alice",   [92, 88, 95, 91, 87]),
    ("Bob",     [78, 82, 75, 80, 85]),
    ("Charlie", [65, 70, 60, 68, 72]),
    ("Diana",   [95, 98, 92, 96, 99]),
    ("Eve",     [55, 48, 62, 51, 58]),
    ("Frank",   [88, 84, 90, 86, 83]),
]

# Create student records
classroom = []
for name, scores in data:
    student = create_student(name, scores)
    classroom.append(student)

# Display full report
display_class_report(classroom)
```

```python
# =====================================================
# PROJECT: Sales Data Analyzer
# =====================================================
import csv
import json
from datetime import datetime

def create_sales_data():
    """Create sample sales CSV"""
    sales = """date,product,category,quantity,price,city
2024-01-05,Laptop,Electronics,2,75000,Mumbai
2024-01-07,Phone,Electronics,5,25000,Delhi
2024-01-08,Desk,Furniture,3,12000,Bangalore
2024-01-10,Laptop,Electronics,1,75000,Mumbai
2024-01-12,Chair,Furniture,8,5000,Delhi
2024-01-15,Tablet,Electronics,4,35000,Chennai
2024-01-18,Phone,Electronics,6,25000,Bangalore
2024-01-20,Sofa,Furniture,2,45000,Mumbai
2024-01-22,Laptop,Electronics,3,75000,Delhi
2024-01-25,Tablet,Electronics,2,35000,Mumbai
2024-01-28,Chair,Furniture,5,5000,Chennai
2024-01-30,Phone,Electronics,3,25000,Delhi
"""
    with open("sales.csv", "w") as f:
        f.write(sales)
    print(" sales.csv created!\n")


def load_sales(filename):
    """Load and process sales data"""
    sales = []
    try:
        with open(filename, "r") as file:
            reader = csv.DictReader(file)
            for row in reader:
                sales.append({
                    "date"    : row["date"],
                    "product" : row["product"],
                    "category": row["category"],
                    "quantity": int(row["quantity"]),
                    "price"   : int(row["price"]),
                    "city"    : row["city"],
                    "revenue" : int(row["quantity"]) * int(row["price"])
                })
    except FileNotFoundError:
        print(" sales.csv not found!")
        return []
    return sales


def sales_summary(sales):
    """Overall summary"""
    total_revenue  = sum(s["revenue"] for s in sales)
    total_quantity = sum(s["quantity"] for s in sales)
    total_orders   = len(sales)
    avg_order      = total_revenue // total_orders

    print("=" * 45)
    print("        SALES ANALYSIS REPORT")
    print("=" * 45)
    print(f"  Total Orders   : {total_orders}")
    print(f"  Items Sold     : {total_quantity}")
    print(f"  Total Revenue  : ₹{total_revenue:,}")
    print(f"  Avg Order Value: ₹{avg_order:,}")


def revenue_by_category(sales):
    """Revenue per category"""
    categories = {}
    for s in sales:
        cat = s["category"]
        categories[cat] = categories.get(cat, 0) + s["revenue"]

    print("\n REVENUE BY CATEGORY")
    print("-" * 35)
    for cat, rev in sorted(categories.items(),
                           key=lambda x: x[1], reverse=True):
        print(f"  {cat:15} : ₹{rev:,}")


def top_products(sales):
    """Best selling products by revenue"""
    products = {}
    for s in sales:
        p = s["product"]
        if p not in products:
            products[p] = {"revenue": 0, "quantity": 0}
        products[p]["revenue"]  += s["revenue"]
        products[p]["quantity"] += s["quantity"]

    print("\n TOP PRODUCTS BY REVENUE")
    print("-" * 40)
    sorted_products = sorted(products.items(),
                             key=lambda x: x[1]["revenue"],
                             reverse=True)
    for product, info in sorted_products:
        print(f"  {product:8} | "
              f"Revenue: ₹{info['revenue']:>10,} | "
              f"Qty: {info['quantity']}")


def revenue_by_city(sales):
    """City-wise performance"""
    cities = {}
    for s in sales:
        c = s["city"]
        cities[c] = cities.get(c, 0) + s["revenue"]

    print("\n  REVENUE BY CITY")
    print("-" * 35)
    max_rev = max(cities.values())
    for city, rev in sorted(cities.items(),
                            key=lambda x: x[1], reverse=True):
        bar_len = int((rev / max_rev) * 15)
        bar = "█" * bar_len
        print(f"  {city:12} | {bar:<15} ₹{rev:,}")


def save_report(sales):
    """Save analysis to JSON"""
    total_rev = sum(s["revenue"] for s in sales)
    report = {
        "generated_on"  : str(datetime.now()),
        "total_orders"  : len(sales),
        "total_revenue" : total_rev,
        "top_product"   : max(sales, key=lambda x: x["revenue"])["product"]
    }
    with open("sales_report.json", "w") as f:
        json.dump(report, f, indent=4)
    print("\n Report saved to sales_report.json!")


# ── Run Everything ──────────────────────────────
create_sales_data()
sales = load_sales("sales.csv")

if sales:
    sales_summary(sales)
    revenue_by_category(sales)
    top_products(sales)
    revenue_by_city(sales)
    save_report(sales)
```
