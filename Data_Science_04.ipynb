```python
# drop missing value
messy_dt = messy_dt.dropna(subset=["customer_name"])
```

```python
messy_dt["age"] = messy_dt["age"].fillna(messy_dt["age"].median())
```

```python
messy_dt["discount"] = messy_dt["discount"].fillna(0)
```

```python
messy_dt["delivery_days"] = messy_dt["delivery_days"].fillna(messy_dt["age"].median())
```

```python
messy_dt["rating"] = messy_dt["rating"].fillna(messy_dt["rating"].median())
```

```python
messy_dt.isnull().sum()
```

```python
messy_dt.shape
```

```python
# Remove Duplicate transaction
messy_dt = messy_dt.drop_duplicates(subset = columns)
```

```python
messy_dt.shape[0]
```

```python
columns = ["customer_name","city","product_category","quantity","unit_price","order_date"]

messy_dt.duplicated(subset = columns).sum()
```

```python
messy_dt[messy_dt.duplicated(subset = columns,keep=False)].value_counts("customer_name")
```

```python
messy_dt = messy_dt.drop(messy_dt[messy_dt["quantity"]<=0].index)
```

```python
messy_dt[messy_dt["quantity"]<=0]
```

```python
messy_dt.loc[messy_dt["age"]<=0, "age"] = messy_dt["age"].median()
```

```python
messy_dt[messy_dt["age"]<=0]
```

```python
messy_dt.loc[messy_dt["age"] > 100, "age"] = messy_dt["age"].median()
```

```python
messy_dt[messy_dt["age"]>100]
```

```python
messy_dt["order_date"] = pd.to_datetime(messy_dt["order_date"])
messy_dt.dtypes
```

```python
print (f"total shape:  {messy_dt.shape[0]}")


print(f" duplicate remaining : {messy_dt.duplicated(subset = columns).sum()}")

print(f" zero quantity row: {messy_dt[messy_dt["quantity"]<=0].shape[0]}")

print(f" invalied age : {messy_dt[(messy_dt["age"]<=0) | (messy_dt["age"]>100 )].shape[0]}")

print(f"data dtype: {messy_dt["order_date"].dtype}")

print(f"total null values: {messy_dt.isnull().sum()}")
```

```python
print(messy_dt.columns.tolist())
```

```python
# Add revenue column.
messy_dt["revenue"] = messy_dt["quantity"] * messy_dt["unit_price"] * (1 - messy_dt["discount"] / 100)
```

```python
messy_dt[["customer_name","quantity","unit_price","discount","revenue"]].head()
```

```python
# product category generates most revenue
revenue_by_category = messy_dt.groupby("product_category")["revenue"].sum()
revenue_by_category
```

```python
most_revenue_category = revenue_by_category.idxmax()
max_revenue_amount = revenue_by_category.max()
```
